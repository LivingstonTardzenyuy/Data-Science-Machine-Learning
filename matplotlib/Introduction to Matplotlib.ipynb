## Matplotlib example Workflow

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary
num_better = {
                "Almond butter": 10,
                "Peanut butter": 8,
                "Cashew butter":6
}
fig, ax = plt.subplots()
ax.bar(num_better.keys(), height=num_better.values());
ax.set(
    title="Dan's Nut butter store",
    xlabel="Butter",
    ylabel="Price ($)"
);
```

```python
tatum_fc = {
    "Serge": 10,
    "Sule": 11,
    "Mustapha":12,
    "Cedric":17,
    "Kiply":23
}
fig,ax= plt.subplots()
ax.bar(tatum_fc.keys(), height=tatum_fc.values())
ax.set(
    title="Tatum FC",
    xlabel="Player Names",
    ylabel="Number of Goals"
);
```
